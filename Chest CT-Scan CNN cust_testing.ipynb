```python
def predict_cancer(img_path):
  img = cv2.imread(img_path)
  img = cv2.resize(img,(224,224))
  pred = model.predict(img.reshape(1,224,224,3))
  label = ["adenocarcinoma","large.cell.carcinoma","normal","squamous.cell.carcinoma"]
  for i in range(len(pred[0])):
    if pred[0][i]==1:
     return label[int(i)]
```

```python
p = predict_cancer("/content/drive/MyDrive/Projects/Chest CT-Scan/Data/train/squamous.cell.carcinoma_left.hilum_T1_N2_M0_IIIa/000004 (7).png")
```
